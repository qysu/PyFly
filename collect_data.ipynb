```python
import pandas as pd

import time
import datetime
from datetime import date

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait 
from selenium.common.exceptions import StaleElementReferenceException
from selenium.webdriver.support import expected_conditions as EC

csv_name = '%02d' % date.today().month + '%02d' % date.today().day
path = r'/Users/kittysu/Documents/Projects/PyFly/Flight_Data/' + csv_name + '.csv'

def search(origin, destination, date, days):
    
    url = "https://www.expedia.com/Flights-Search?trip=oneway&leg1=from:{0},to:{1},departure:{2}TANYT&passengers=adults:1,children:0,seniors:0,infantinlap:Y&options=cabinclass%3Aeconomy&mode=search&origref=www.expedia.com".format(origin,destination,date)
    
    driver.get(url)
    driver.implicitly_wait(100)

    df = pd.DataFrame()
        
    # Scrap required data
    scrap = WebDriverWait(driver,3).until(EC.visibility_of_all_elements_located((By.XPATH, "//span[@data-test-id='departure-time']")))
    departures = [val.text for val in scrap]
        
    scrap = WebDriverWait(driver,3).until(EC.visibility_of_all_elements_located((By.XPATH, "//span[@data-test-id='arrival-time']")))
    arrivals = [val.text for val in scrap]
        
    scrap = WebDriverWait(driver,3).until(EC.visibility_of_all_elements_located((By.XPATH, "//span[@data-test-id='listing-price-dollars']")))
    airfares = [val.text for val in scrap]
        
    scrap = WebDriverWait(driver,3).until(EC.visibility_of_all_elements_located((By.XPATH, "//span[@data-test-id='airline-name']")))
    airlines = [val.text for val in scrap]
        
    scrap = WebDriverWait(driver,3).until(EC.visibility_of_all_elements_located((By.XPATH, "//span[@data-test-id='duration']")))
    durations = [val.text for val in scrap]
        
    scrap = WebDriverWait(driver,3).until(EC.visibility_of_all_elements_located((By.XPATH, "//span[@class='number-stops']")))
    num_stops = [val.text for val in scrap]
        
    # Combine scrapped data
    for i in range(len(departures)):
        try:
            df.loc[i, 'departure_time'] = departures[i]
        except Exception as e:
            pass
        try:
            df.loc[i, 'arrival_time'] = arrivals[i]
        except Exception as e:
            pass
        try:
            df.loc[i, 'airfare'] = airfares[i]
        except Exception as e:
            pass
        try:
            df.loc[i, 'airline'] = airlines[i]
        except Exception as e:
            pass
        try:
            df.loc[i, 'duration'] = durations[i]
        except Exception as e:
            pass
        try:
            df.loc[i, 'num_stops'] = num_stops[i]
        except Exception as e:
            pass
        
    df['departure_date'] = date
    df['days_to_departure'] = days
    
    # Save to csv
    if days==1:
        export_csv = df.to_csv (path, index = None, header=True) 
    else:
        df.to_csv(path + csv_name + '.csv', mode='a', index=False, header=False, sep=",")
    
    return
    
if __name__ == "__main__":
    
    origin = 'YYZ'
    destination = 'SYD' 
    
    driver = webdriver.Chrome()
    
    purchase_date = date.today()
    delta = datetime.timedelta(days=1)

    curr_departure_date = date.today() + datetime.timedelta(days=1)
    final_departure_date = date.today() + datetime.timedelta(days=31)
          
    while curr_departure_date <= final_departure_date:
        
        flight_date = curr_departure_date.strftime('%m/%d/%Y')
        days_to_departure = (curr_departure_date-purchase_date).days
    
        search(origin, destination, flight_date, days_to_departure)
        
        curr_departure_date += delta
    
    driver.close()
    
    print("done")
```
